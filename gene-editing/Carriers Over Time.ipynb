```python
colors = itertools.cycle(['r', 'g', 'b', 'k'])
markers = itertools.cycle(['o', 'v', 's', 'd'])
xlabels = df.year.unique()

fig = plt.figure(figsize=(16, 12), dpi=300, facecolor='white')
plt.title('Proportion of genotyped bulls and cows carrying known recessives, 2000-2015',
         weight='bold')
ax = fig.add_subplot(111)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)  
ax.set_xlabel('Birth year')
ax.set_ylabel('Proportion (%)')

ax.plot(df['year'][(df['sex'] == 'F') & 
                          (df['recessives'] == 0)],
        df['count_ratio'][(df['sex'] == 'F') & 
                          (df['recessives'] == 0)],
        label='Cows carrying no known recessives',
        marker=markers.next(),
        c=colors.next()
       )

ax.plot(df['year'][(df['sex'] == 'M') & 
                          (df['recessives'] == 0)],
        df['count_ratio'][(df['sex'] == 'M') & 
                          (df['recessives'] == 0)],
        label='Bulls carrying no known recessives',
        marker=markers.next(),
        c=colors.next())

ax.plot(df['year'][(df['sex'] == 'F') & 
                          (df['recessives'] == 1)],
        df['count_ratio'][(df['sex'] == 'F') & 
                          (df['recessives'] == 1)],
        label='Cows carrying one or more known recessives',
        marker=markers.next(),
        c=colors.next())

ax.plot(df['year'][(df['sex'] == 'M') & 
                          (df['recessives'] == 1)],
        df['count_ratio'][(df['sex'] == 'M') & 
                          (df['recessives'] == 1)],
        label='Bulls carrying one or more known recessives',
        marker=markers.next(),
        c=colors.next())

ax.set_ylim(0.0, 100.0)

# Turn off the top and right tick marks
plt.tick_params(
    axis='both',       # changes apply to both axes
    which='both',      # both major and minor ticks are affected
    right='off',
    labelright='off',
    top='off',         # ticks along the top edge are off
    labeltop='off')    # labels along the bottom edge are off

# Place the legend
ax.legend(loc='best')

# Annotate with source, etc.
#plt.text(0, 0, "Data source: Council on Dairy Cattle Breeding"
       "\nAuthor: John B. Cole (AGIL, ARS, USDA / john.cole@ars.usda.gov)")

plt.tight_layout(pad=1., w_pad=0.5, h_pad=0.95)
plt.show()

filename = 'holstein-carriers-by-year.png'
fig.savefig(filename, dpi=300)
```
