```python
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.mpl_style', 'default')
import statsmodels.api as sm
import itertools
```

First, let's load the results from the small model of polled included in the default settings. This involves loading four animal files (live cows, dead cows, live bulls, and dead bulls). We will load them and merge them into a single data frame.

```python
for sim in xrange(1,11):
    # Load the individual history files
    lc = pd.read_csv('low_low/%s/cows_history_ran_low_low_20.txt'%sim, sep='\t', header=0)
    dc = pd.read_csv('low_low/%s/dead_cows_history_ran_low_low_20.txt'%sim, sep='\t', header=0)
    lb = pd.read_csv('low_low/%s/bulls_history_ran_low_low_20.txt'%sim, sep='\t', header=0)
    db = pd.read_csv('low_low/%s/dead_bulls_history_ran_low_low_20.txt'%sim, sep='\t', header=0)
    inbreeding = pd.read_csv('low_low/%s/pedigree_20.txt.solinb'%sim, delim_whitespace=True,
                             skipinitialspace=True, names=['animal','inbreeding'])
    # Stack the individual animal datasets
    allan = lc.append(dc.append(lb.append(db)))
    # Merge in the coefficients of inbreeding (Pandas defaults to an inner join)
    all_animals = pd.merge(allan, inbreeding, on='animal')
    all_animals['rep'] = sim
    if sim == 1:
        all_ran_replicates = all_animals
    else:
        all_ran_replicates = pd.concat([all_ran_replicates, all_animals])
```

```python
for sim in xrange(1,11):
    # Load the individual history files
    lc = pd.read_csv('low_low/%s/cows_history_toppct_low_low_20.txt'%sim, sep='\t', header=0)
    dc = pd.read_csv('low_low/%s/dead_cows_history_toppct_low_low_20.txt'%sim, sep='\t', header=0)
    lb = pd.read_csv('low_low/%s/bulls_history_toppct_low_low_20.txt'%sim, sep='\t', header=0)
    db = pd.read_csv('low_low/%s/dead_bulls_history_toppct_low_low_20.txt'%sim, sep='\t', header=0)
    inbreeding = pd.read_csv('low_low/%s/pedigree_20.txt.solinb'%sim, delim_whitespace=True,
                             skipinitialspace=True, names=['animal','inbreeding'])
    # Stack the individual animal datasets
    allan = lc.append(dc.append(lb.append(db)))
    # Merge in the coefficients of inbreeding (Pandas defaults to an inner join)
    all_animals = pd.merge(allan, inbreeding, on='animal')
    all_animals['rep'] = sim
    if sim == 1:
        all_toppct_replicates = all_animals
    else:
        all_toppct_replicates = pd.concat([all_toppct_replicates, all_animals])
```

```python
# We have 10 relicates for each simulation
for sim in xrange(1,11):
    # Load the individual history files
    lc = pd.read_csv('low_low/%s/cows_history_pryce_low_low_20.txt'%sim, sep='\t', header=0)
    dc = pd.read_csv('low_low/%s/dead_cows_history_pryce_low_low_20.txt'%sim, sep='\t', header=0)
    lb = pd.read_csv('low_low/%s/bulls_history_pryce_low_low_20.txt'%sim, sep='\t', header=0)
    db = pd.read_csv('low_low/%s/dead_bulls_history_pryce_low_low_20.txt'%sim, sep='\t', header=0)
    inbreeding = pd.read_csv('low_low/%s/pedigree_20.txt.solinb'%sim, delim_whitespace=True,
                             skipinitialspace=True, names=['animal','inbreeding'])
    # Stack the individual animal datasets
    allan = lc.append(dc.append(lb.append(db)))
    # Merge in the coefficients of inbreeding (Pandas defaults to an inner join)
    all_animals = pd.merge(allan, inbreeding, on='animal')
    all_animals['rep'] = sim
    if sim == 1:
        all_pryce_replicates = all_animals
    else:
        all_pryce_replicates = pd.concat([all_pryce_replicates, all_animals])
# Print first few lines of dataframe
#all_animals.head()
```

```python
# Now load the Pryce+recessives data so that we can compare EBV.
# We have 10 relicates for each simulation
for sim in xrange(1,11):
    # Load the individual history files
    lc = pd.read_csv('low_low/%s/cows_history_pryce_r_low_low_20.txt'%sim, sep='\t', header=0)
    dc = pd.read_csv('low_low/%s/dead_cows_history_pryce_r_low_low_20.txt'%sim, sep='\t', header=0)
    lb = pd.read_csv('low_low/%s/bulls_history_pryce_r_low_low_20.txt'%sim, sep='\t', header=0)
    db = pd.read_csv('low_low/%s/dead_bulls_history_pryce_r_low_low_20.txt'%sim, sep='\t', header=0)
    inbreeding = pd.read_csv('low_low/%s/pedigree_20.txt.solinb'%sim, delim_whitespace=True,
                             skipinitialspace=True, names=['animal','inbreeding'])
    # Stack the individual animal datasets
    allan_r = lc.append(dc.append(lb.append(db)))
    # Merge in the coefficients of inbreeding (Pandas defaults to an inner join)
    all_animals_r = pd.merge(allan_r, inbreeding, on='animal')
    all_animals_r['rep'] = sim
    if sim == 1:
        all_pryce_replicates_r = all_animals_r
    else:
        all_pryce_replicates_r = pd.concat([all_pryce_replicates_r, all_animals_r])
# Print first few lines of dataframe
#all_animals.head()
```

Random Mating
====
Causes of death (A=culled for age, N=culled for population size, R=died from recessive)

```python
ct_ran = pd.crosstab(all_ran_replicates['died'], all_ran_replicates['cause']).apply(lambda r: r/r.sum(), axis=1)
ct_ran
```

Truncation Selection
====
Causes of death (A=culled for age, N=culled for population size, R=died from recessive)

```python
ct_toppct = pd.crosstab(all_toppct_replicates['died'], all_toppct_replicates['cause']).apply(lambda r: r/r.sum(), axis=1)
```

Pryce's Algorithm
====
Causes of death (A=culled for age, N=culled for population size, R=died from recessive)

```python
ct_pryce = pd.crosstab(all_pryce_replicates['died'], all_pryce_replicates['cause']).apply(lambda r: r/r.sum(), axis=1)
```

Modified Pryce's Algorithm
====
Causes of death (A=culled for age, N=culled for population size, R=died from recessive)

```python
ct_pryce_r = pd.crosstab(all_pryce_replicates_r['died'], all_pryce_replicates_r['cause']).apply(lambda r: r/r.sum(), axis=1)
```
